````python
# Copy and run generated codes here

# Unsupervised Machine Learning Demo: Clustering with K-Means

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# --- 1. Generate synthetic data (2D for visualization) ---
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.7, random_state=42)

# --- 2. Visualize true clusters ---
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y_true, cmap='viridis', s=40)
plt.title("True Clusters")
plt.xlabel("Feature 1"); plt.ylabel("Feature 2")

# --- 3. Run clustering algorithm (K-Means) ---
kmeans = KMeans(n_clusters=3, random_state=42)
y_pred = kmeans.fit_predict(X)

# --- 4. Visualize predicted clusters ---
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', s=40, marker='x')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', marker='*', s=200, label='Centroids')
plt.title("Predicted Clusters (K-Means)")
plt.xlabel("Feature 1"); plt.ylabel("Feature 2")
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# ============================================================
# Unsupervised Machine Learning Demo: Cluster Analysis
# ============================================================
# This script generates a synthetic dataset, visualizes it,
# performs clustering using K-means, and visualizes the results.
# Users can adjust parameters like number of clusters, noise, etc.
# ============================================================

# --- Import required libraries ---
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# ============================================================
# 1. Generate synthetic dataset
# ============================================================

# PARAMETERS (users can adjust these)
n_samples = 300          # total number of data points
n_features = 2           # number of features (2D for easy visualization)
n_clusters = 3           # number of clusters to generate
cluster_std = 0.7        # standard deviation of clusters -> higher = more overlap (higher difficulty)
random_state = 42        # for reproducibility

# Generate synthetic data
X, y_true = make_blobs(
    n_samples=n_samples,
    n_features=n_features,
    centers=n_clusters,
    cluster_std=cluster_std,
    random_state=random_state
)

# ============================================================
# 2. Visualize the true (actual) clusters
# ============================================================

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.title("True Cluster Membership (Ground Truth)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

# Scatter plot with true labels
for i in range(n_clusters):
    plt.scatter(X[y_true == i, 0], X[y_true == i, 1], label=f"Cluster {i+1}", s=50)

plt.legend()
plt.grid(True)

# ============================================================
# 3. Apply clustering algorithm (K-Means)
# ============================================================

# Initialize and fit the KMeans model
kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)
y_pred = kmeans.fit_predict(X)

# ============================================================
# 4. Visualize predicted clusters vs true clusters
# ============================================================

plt.subplot(1, 2, 2)
plt.title("Predicted Clusters (K-Means Results)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

# Plot data points with predicted cluster assignment
for i in range(n_clusters):
    plt.scatter(X[y_pred == i, 0], X[y_pred == i, 1], label=f"Predicted Cluster {i+1}", s=50, marker='x')

# Plot cluster centers
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='black', marker='*', s=200, label='Centroids')

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ============================================================
# Notes:
# - Increase 'cluster_std' to make clusters overlap more (harder task).
# - Decrease 'cluster_std' for clearly separable clusters (easier task).
# - Try changing 'n_clusters' to see how K-Means behaves when the
#   assumed number of clusters doesn't match the true number.
# ============================================================
````

While the above code is enough to solve the task, examining the resulting clustering is suitable is more challenging. 

Since we use synthetic data, an intuitive approach is to compare the true cluster membership (i.e., the variable `labels` that the `make_blobs` function returns) to the output of the `fit_predict` method. Unfortunately, this does not work because of the **arbitrary cluster indexing** in both `make_blobs` and K-means. While `make_blobs` assigns indices to clusters when generating synthetic data, K-means does not know these labels and assigns its own arbitrary indices based on how it organizes clusters during training.

This discrepancy means that even if the clustering is perfectly accurate in terms of data point grouping, the indices of clusters in `make_blobs` and K-means may not align (e.g., what `make_blobs` labels as cluster `0`, K-means might label as cluster `2`). 

To align indices for accurate comparison, we can use the **Hungarian method** to optimally map K-means cluster indices to the true labels, allowing us to objectively assess clustering performance without being affected by differing label assignments.
